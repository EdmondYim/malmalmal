```python
import json
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# JSON 파일 로드
with open("kw_weighting_results\class_keywords.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 폰트 (한글 깨짐 방지)
font_path = "NanumGothic.ttf"

# 각 클래스별 워드클라우드 생성
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for idx, (cls, words) in enumerate(data.items()):
    text = " ".join(words)
    wc = WordCloud(
        font_path=font_path,
        background_color="white",
        width=600,
        height=400
    ).generate(text)

    axes[idx].imshow(wc, interpolation="bilinear")
    axes[idx].axis("off")
    axes[idx].set_title(cls, fontsize=14)

plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import seaborn as sns

for cls, words in data.items():
    counts = Counter(words)
    common = counts.most_common(10)
    sns.barplot(x=[w for w, _ in common], y=[c for _, c in common])
    plt.title(cls)
    plt.xticks(rotation=45)
    plt.show()
```
